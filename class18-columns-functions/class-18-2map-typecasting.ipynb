```python
#df1.Total = 500 #  not add new column
df1['Total'] = 500 # add new column
# del df1.Total  #  not delete column
#df1.toal = 300 # update column

df1
```

```python
df1['obtained'] = df1['s1'] + df1['s2'] +df1['s3'] + df1['s4'] + df1['s5']
df1
```

```python
df1['Percentage'] = df1['obtained'] / df1['Total'] * 100
df1
```

```python
def grade(per : int) -> str:
    if per >= 80:
        return "A+"
    elif per >= 70:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 50:
        return "C"
    elif per >= 40:
        return "D"
    elif per >= 33:
        return "E"
    else:
        return "F"

grade (82)
grade (30)
```

```python
df1['Percentage'].apply(grade)
```

```python
df1['Grade'] = df1['Percentage'].apply(grade)
df1
```

```python
print (df1)
```

```python
from typing import Dict

remarks : Dict[str, str] = {
    "A+": "Excellent",
    "A": "Very Good",
    "B": "Good",
    "C": "Average",
    "D": "Below Average",
    "E": "Poor",
    "F": "Very Poor"
}
```

```python
df1['Remarks'] = df1.Grade.map(remarks)
df1
```
