```python
# Define the function
def fn_obtained(s1: int, s2: int, s3: int, s4: int, s5: int) -> tuple[int, int, float, str]:
    total: int = 500
    obtained: int = s1 + s2 + s3 + s4 + s5
    percentage: float = obtained / total * 100
    grade: str = ""
    if percentage >= 80:
        grade = "A+"
    elif percentage >= 70:
        grade = "A"
    elif percentage >= 60:
        grade = "B"
    elif percentage >= 50:
        grade = "C"
    elif percentage >= 40:
        grade = "D"
    elif percentage >= 33:
        grade = "E"
    else:
        grade = "F"
    return total, obtained, percentage, grade

# Apply the function and expand the results into separate columns
results = df1[['s1', 's2', 's3', 's4', 's5']].apply(lambda x: fn_obtained(*x), axis=1, result_type='expand')
results.columns = ['Total', 'Obtained', 'Percentage', 'Grade']

# Concatenate the results with the original DataFrame
df1 = pd.concat([df1, results], axis=1)

print(df1)
```
